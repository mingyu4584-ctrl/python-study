```python
# 예제
# 네이버 메인 페이지 title 가져오기

import requests

from bs4 import BeautifulSoup

# 3. 크롤링 할 웹 페이지 주소
url = 'https://www.naver.com'

# 4. 해당 url로 Get 요청을 보내고 응답을 받음
response = requests.get(url)

# 5. 서버 응답이 정상인지 확인 (200 이면 성공)
print(response.status_code)

# 6. 받아온 HTML 내용을 BeaulifulSoup 객체로 변환
# 'html_parser'는 파이썬 기본 HTML 분석기
soup = BeautifulSoup(response.text,'html.parser')

# 7. HtML 문서의 <title> 태그를 찾음
title = soup.title

# 8. title 태그 안의 텍스트만 출력
print(title.text)

```

```python
import json

# 크롤링 할 상품 페이지
url ="https://search.naver.com/search.naver?where=news&query=인공지능"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# 페이지 요청
response = requests.get(url, headers=headers)
html= response.text
soup = BeautifulSoup(html, 'html.parser')

# __NEXT_DATA__ JSON 꺼내기
links = soup.find_all("a", href=True)

count = 0

# # productName 찾기
for a in links:
    href = a["href"].strip()
    
    title = a.get_text(strip=True)

    if (title and href.startswith("https://")
        and "view" in href
        ):
        count += 1
        print(f"{count}. {title}")
        print(f"    {href})")
```

```python
from selenium import webdriver  # 
from selenium.webdriver.common.by import By #
from selenium.webdriver.common.keys import Keys # 
import time # 대기 시간 제어용
import csv  # csv 파일 저장용

# 크롬 실행 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')   # 브라우저 크기 설정함
options.add_argument(
    '--disable-blink-features=AutomationControlled')   # 자동화 탐지 방지

driver = webdriver.Chrome(options=options)  # 크롬 브라우저 실행

driver.get('https://www.naver.com') # 네이버 메인 접속
time.sleep(2)

search_box = driver.find_element(By.CSS_SELECTOR, "input#query")    # 검색창 찾기
search_box.click()  # 검색창 클릭
time.sleep(0.5) # 클릭 후 잠시 대기

keyword = "인공지능"    # 검색 할 키워드
for char in keyword:    # 키워드를 한 글자씩 반복
    search_box.send_keys(char)  # 검색창에 글자 입력
    time.sleep(0.25)    # 타이핑 딜레이 (사람처럼 보이게)

search_box.send_keys(Keys.ENTER)    # 엔터 키를 입력 -> 검색 실행
time.sleep(2)   # 검색 결과 로딩 대기

news_tab = driver.find_element(By.LINK_TEXT, "뉴스")    # 상단 탭 중 "뉴스" 찾기
news_tab.click()    # 뉴스 탭 클릭
time.sleep(3)   # 뉴스 페이지 로딩 대기

print("뉴스 검색 페이지 도착")  # 진행 확인용


news_list = driver.find_elements(By.CSS_SELECTOR,"a")   # 페이지 내 모든 a 태그 수집
news_data = []  # 뉴스 데이터 저장 리스트


# count = 0   # 뉴스 갯수 카운트 변수 선언 (데이터 몇개 있는지 확인용)

for a in news_list: # 모든 a 하나씩 검사
    title = a.text.strip()  # a 태그 안의 텍스트(뉴스 제목)
    href = a.get_attribute("href")  # a 태그의 링크 주소

    if (title   # 제목이 존재하고
        and href    # 링크가 존재 하고
        and href.startswith("https://") # 실제 외부 URL 이며
        and ("view" in href or "article" in href)   # 기사 URL 형태일 때
        ):
        # count += 1  # 뉴스 번호 증가
        # print(f"{count}. {title}")  # 뉴스 제목 출력
        # print(f"    {href})")   # 뉴스 링크 출력
        news_data.append([title,href])
print("총 뉴스 개수:", len(news_data))   # 수집 된 뉴스 중 개수 출력

with open("naver_news.csv","w",newline="",encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["제목","링크"])
    writer.writerows(news_data)

print("저장완료")
# driver.quit() # 브라우저 종료 (필요시에 주석 해제)
```
