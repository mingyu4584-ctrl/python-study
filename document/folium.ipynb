```python
import folium

m = folium.Map(location=[36.087600969505985,129.37897836553745],zoom_start=15,
               radius=100, color='#34ebc6',fill_color='#c634eb')
folium.Marker([36.087600969505985,129.37897836553745],popup='내집',tooltip='내집').add_to(m)
m.add_child(folium.ClickForMarker(popup='내가 클릭한 곳'))
m.add_child(folium.LatLngPopup())
m
```
