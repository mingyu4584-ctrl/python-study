```python
# 그래프를 그리기 위한 matplotib 모듈 불러오기
import matplotlib.pylab as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]
y2 = [1,4,9,16,25]

# x와 y값으로 선 그래프를 그리고 화면에 표시
plt.plot(x, y)

# 제목 붙이기
plt.title("My First Graph")

# 축 이름 붙이기
plt.xlabel("X")
plt.ylabel("Y")

# 범례 추가
plt.plot(x,y, label="Linear")
plt.plot(x,y2, label="Square")

# 선 색상 바꾸기
plt.plot(x,y, color="red", label="Red Line")

# 선에 마커 표시
plt.plot(x,y,marker="o",label="Dot Line")

# 격자 표시
plt.grid()

# 범례 추가된 것 적용(범례 표시)
plt.legend()

# 그래프 보여주기
plt.show()
```

```python
# 가장 많이 쓰는 기본 형태(암기)
import matplotlib.pylab as plt

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y, marker="o",label="Example")
plt.title("Matplotlib Basic Examples")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.legend()
plt.show()

# plt.plot() -> 그래프 그리기
# label -> 범례 이름
# plt.legend() -> 범례 표시
# plt.grid() -> 눈금
# plt.show() -> 화면 출력
```

##### 실습1. 선 그래프 그리기
###### 아래 데이터를 이용해서 예시와 같이 선 그래프를 그리기

```python
import matplotlib.pylab as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months,sales_2019,color="b",label='2019')
plt.plot(months,sales_2020,color="orange",label='2020')
plt.title("Monthly Sales Comparison (2019~2020)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

##### 실습2. 막대 그래프 그리기
###### 아래 데이터를 바탕으로 예시와 같이 막대 그래프를 그리기

```python
import matplotlib.pylab as plt

categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20,35,15,27,45]

plt.bar(categories,data,width=0.7,color=None,label=None)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(alpha=0.1)
plt.xticks(rotation=45)
plt.ylim(0,50)
plt.legend()
plt.show()
```

##### 실습3. 히스토그램 그리기
###### 임의로 1~6 사이의 주사위 눈 100개를 생성해서 각 숫자가 얼마나 자주 나왔는지 히스토그램 그리기

```python
import matplotlib.pylab as plt
import numpy as np

data = np.random.randint(1,7,100)
bins = np.arange(0.5,7.5)

plt.hist(data, bins, color='skyblue',edgecolor='black',rwidth = 0.7)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.title("주사위 눈 빈도 히스토그램")
plt.xlabel("주사위 눈")
plt.ylabel("빈도")
plt.xticks(np.arange(1,7))
plt.grid(axis='y',linestyle=':',alpha=0.5)
plt.legend()
plt.locator_params(axis='y',integer = True)
plt.show()
```

##### 실습4. 산점도 그리기
###### x값은 0부터 10까지 30개의 균등한 간격으로 구성된 수열을 만들고, y값은 y = 3x + 2 + 난수(정규분포, 평균 0, 표준편자 3)로 생성하여 산점도 그리기

```python
import matplotlib.pylab as plt
import numpy as np

np.random.seed()

x = np.linspace(0,10,30)
n = np.random.normal(0,3,30)
y = 3*x + 2 + n
# y_noisy = y +n.uniform(n,size=30)

plt.scatter(x,y,color='purple',marker='o')

plt.title("노이즈가 추가된 산점도")
plt.xlabel("x 값")
plt.ylabel("y 값")
plt.grid(True, linestyle=':', alpha=0.5)
plt.legend()
plt.show()
```

##### 실습5. 파이 차트 그리기
###### 아래와 데이터를 이용해서 예시와 같은 파이 차트 그리기

```python
import matplotlib.pylab as plt

labels = ['Apple','Banana','Mango','Blueberry']
sizes = [15,30,45,10]
explode = (0,0.1,0,0)

plt.pie(sizes, labels = labels, autopct='%1.1f%%', explode = explode, startangle = 90,
        shadow = True, wedgeprops = {'edgecolor':'black'})
plt.axis("equal")
plt.show()
```
