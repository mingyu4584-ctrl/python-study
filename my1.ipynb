```python
# 파일을 쓰기모드 (w)로 열고 파일 객체를 f에 저장합니다. f == data.txt
f = open("data.txt",'w',encoding='utf-8')

# 파일에 문자열을 씁니다.
f.write(f"파이썬 파일 입출력 예제 입니다. \n")
f.write(f"이 파일은 write()로 작성되었습니다. \n")

# 파일을 닫아 자원을 해제하고, 내용이 안전하게 저장되도록 합니다.
f.close()
```

```python
# try exept 가장 기본 예제
try:    # 예외가 발생 할 수 있는 코드
    num = int(input("숫자를 입력하세요:"))
    print("입력한 숫자:",num)   # 에러시 출력 안됨
except:  # 예외가 발생했을 때 실행할 코드
    print("숫자가 아닌 값을 입력했습니다.")

# int() 변환 실패 -> ValueError
# 프로그램이 바로 죽지 않고 메세지 출력
```

```python
try:
    a = int(input("첫번째 숫자:"))
    b = int(input("두번째 숫자:"))
    result = a/b
    print("결과:",result)

except ValueError:
    print("숫자만 입력해야 합니다.")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
# else:   # try가 성공했을 때만 실행
#     print("입력에 성공 하였습니다.")
finally:    # 에러가 나던 안 나던 항상 실행
    print("언제든")
```

```python
r_age = 0

while True: # try가 성공 할 때 까지 실행 됨 (계속 반복 성공시 까지)
    try:
        age = int(input("나이를 입력하세요:"))
        r_age = age
        print("나이:",r_age)
        break   # 정상적인 숫자 입력시 종료
    
    except ValueError:
        print("숫자로 다시 입력하세요.")    # 다시 while문 실행
```

```python
# 사용자 정의 예외 발생 (조금 어렵습니다)
try:
    age = int(input("나이 입력:"))
    if age < 0:
        raise ValueError("나이는 음수가 될 수 없습니다.\n")
    if age > 100:
        raise ValueError("나이는 음수가 될 수 없습니다.\n")
    print("나이:", age)

except ValueError as e:
    print("에러 발생\n", e)
```
