```python
print(np.zeros((2,3))) # 2행 3열 모두 0.0
print(np.ones((3,2)))  # 3행 2열 모두 1
print(np.empty((2,2))) # 값이 초기화되지 않음
print(np.full((2,2),7))    # [[7 7]
                           #  [7 7]]
print(np.eye(3))    # n x n 단위행렬
print(np.eye(3,5,k=1)) # 3 x 5 행렬에서 k=1인 대각행렬
```

```python
print(np.arange(0,10,2))    # 0 이상 10 미만의 정수 배열을 2간격
print(np.linspace(0,1,5))   # 0 부터 1까지의 균일 간격으로 5개 요소의 배열
```

```python
print(np.random.rand(2,2))  # 2 x 2 크기의 배열 난수로 생성 (0~1사이의 난수)
print(np.random.randn(2,2)) # 2 x 2 크기의 정규분포를 따르는 난수 생성(평균:0 분산:1)
print(np.random.randint(0,10,(2,3)))    # 0~9 사이 정수 2x3 배열
np.random.seed(49)  # 시드 설정
print(np.random.rand(3))    # 항상 동일한 결과
```

```python
from numpy.random import default_rng

rng = default_rng(seed=42)  # rng 객체 생성
print(rng.integers(1,10,size=5))    # 정수 난수
print(rng.random(3))    # 실수 난수
print(rng.normal(0,1,size=(2,3)))   # 정규분포
print(rng.uniform(5,10,size=3)) # 균등분포
```

##### 실습2. 인덱싱과 슬라이싱(1)

##### 실습2. 인덱싱과 슬라이싱(2)
###### 4x5 배열에서 가운데 2x3 부분을 슬라이싱한 뒤 copy()를 이용해 독립 배열

```python
arr = np.arange(1,21).reshape(4,5)
arr1 = arr[1:3,1:4].copy()
print(arr)
print(arr1)
```

###### 3x4 배열에서 짝수이면서 10 이상인 값만 선택(& 활용)

```python
arr = np.array([[4,9,12,7],[10,15,18,3],[2,14,6,20]])
print(arr[(arr >= 10) & (arr%2==0)])
```

###### 5x5 배열에서 2,4번째 행을 선택, 선택된 행에서 열 순서를 [4,0,2]로 재배치

```python
arr = np.arange(1,26).reshape(5,5)
row = [1,3]
col = [4,0,2]
arr1 = arr[row,:]
arr2 = arr1[:,col]
print(arr2)
```

##### 실습2. 인덱싱과 슬라이싱(3)
###### 5x3 배열에서 각 행의 첫 번째 값이 50 이상인 행만 Boolean Indexing으로 선택

```python
arr = np.array([[10,20,30],[55,65,75],[40,45,50],[70,80,90],[15,25,35]])
mask = arr[:,0] >= 50
print(mask)
print(arr[mask])
```

###### 4x4 배열에서 (0,1),(1,3),(2,0),(3,2) 위치의 요소를 한 번에 선택

```python
arr = np.arange(1,17).reshape(4,4)
row = [0,1,2,3]
col = [1,3,0,2]
print(arr[row,col])
```

###### 3차원 배열(2,3,4)에서 모든 블록에서 두 번째 열만 추출해 새로운 2차원 배열 (2,3)을 생성

###### 0~9 난수로 이루어진(4,4)배열을 생성하고, 각각 인덱싱으로 추출(for 이용)

```python
arr = np.random.randint(0,10,(4,4))
print(arr)
print("주대각선 요소:",np.diag(arr))
print("부대각선 요소:",np.diag(np.fliplr(arr)))
```

##### 실습3. NumPy 종합 연습(4)
###### 1부터 50까지의 난수로 된 5x6 배열을 만들고, 배열에서 짝수만 선택하여 출력

```python
arr = np.random.randint(1,51,(5,6))
print(arr[arr%2==0])
```

###### 0부터 99까지의 정수로 이루어진 (10,10) 배열을 생성한 후, [1,3,5]번째 항과 [2,4,6]번째 열의 교차하는 원소들만 선택하여 출력

```python
arr = np.arange(100).reshape(10,10)
row = [1,3,5]
col = [2,4,6]
arr1 = arr[row,:]
arr2 = arr1[:,col]
print(arr2)
```

###### 0~9난수로 이루어진 1차원 배열(길이15)을 생성하고, 짝수 인덱스 위치에 있는 값들 중에서 5 이상인 값만 선택해 출력

```python
arr = np.random.randint(0,10,15)
ar = arr[::2]
ar1 = ar[ar>=5]
print("짝수 인덱스 값들:",ar)
print("짝수 인덱스 중 5 이상인 값들:",ar1)
```
