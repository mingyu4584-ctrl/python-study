### try / except 연습문제(1)


##### 숫자 입력
###### 정수를 입력받아 출력하시오.
###### 숫자가 아니면 "숫자만 입력하세요" 출력.

```python
while True:
    try:
        num = int(input("숫자를 입력하세요:"))
        print(num)
        break
    except ValueError:
        print("숫자만 입력하세요.")
```

##### 나이 입력
###### 나이를 입력받아 출력하시오.
###### 숫자가 아니면 예외 처리.

```python
while True:
    try:
        age = int(input("나이를 입력하세요:"))
        if 0 < age <= 120:
            print(age)
            break
        elif age < 0:
            print("나이는 음수가 될 수 없습니다.")
        else:
            print("비정상적인 나이 입니다.")
    except ValueError:
        print("숫자만 입력하세요.")
```

##### 두 수 나누기
###### 두 정수를 입력받아 나눈 결과를 출력하시오.
###### 숫자 아님
###### 0으로 나누기
###### 를 각각 예외 처리하시오.

```python
while True:
    try:
        num1 = int(input("첫번째 숫자: "))
        num2 = int(input("두번째 숫자: "))
        result = num1/num2
        print(result)
        break
    except ValueError:
        print("숫자를 입력하시오.")
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")
```

##### 음수 나이 금지
###### 나이를 입력받는다.
###### 나이가 음수면 raise ValueError로 에러 발생시키시오.

```python
try:
    age = int(input("나이를 입력"))
    if age < 0:
        raise ValueError("나이가 음수입니다.")
    print("나이:",age)

except Exception as e:
    print("에러발생. \n", e)
```

##### 점수 입력
###### 점수를 입력받아 출력하시오.
###### 정수가 아니면 예외 처리.

```python
try:
    score = int(input("점수 입력"))
    if score < 0:
        raise ValueError("정수가 아닙니다.")
    print("점수:", score)
except Exception as e:
    print("에러 발생.",e)
```

##### 반복 입력
###### 숫자를 입력받는다.
###### 숫자가 입력될 때까지 계속 입력받으시오. (try/except + while)

```python
while True:
    try:
        num = int(input("숫자입력"))
        print("숫자:",num)
        break
    except ValueError:
        print("숫자만 입력")
```

##### 리스트 인덱스 접근
###### 미리 주어진 리스트: nums = [10, 20, 30]
###### 인덱스를 입력받아 해당 값을 출력하시오.
###### 범위를 벗어나면 예외 처리.

```python
num = [10,20,30]
while True:
    try:
        index = int(input("인덱스값"))
        print(num[index])
        break
    except IndexError:
        print("범위를 벗어남")
    except ValueError:
        print("숫자 입력")
```

##### 문자열 → 정수 변환
###### 문자열 리스트: data = ["10", "20", "a", "30"]
###### 각 요소를 정수로 변환하여 출력하시오.
###### 변환 실패 시 예외 처리 후 건너뛰기.

```python
data = ['10','20','a','30']
new_data = []
for i in data:
    try:
        new_data.append(int(i))
    except ValueError:
        pass
print(new_data)
```

##### 사용자 정의 에러
###### 입력받은 숫자가 100보다 크면
###### 직접 ValueError를 발생시키시오.

```python
while True:
    try:
        num = int(input("숫자 입력"))
        if num > 100:
            raise ValueError("100 이상의 숫자입니다.")
        print(num)
        break
    except Exception as e:
        print("에러발생.",e)
```

##### 모든 예외 처리
###### 숫자 하나를 입력받아
###### 어떤 에러든 발생하면 "에러 발생" 출력.

```python
try:
    num = int(input("숫자입력"))
    print(num)
except ValueError:
    print("에러발생.")
```

### 파일 읽기 / 저장하기 연습문제

##### 파일 읽기 기본
###### data.txt 파일을 열어 내용을 출력하시오.
###### 파일이 없으면 예외 처리.

```python
try:
    with open('data.txt','r',encoding='utf-8') as f:
        print(f.read())

except FileNotFoundError:
    print("파일이 없음")
```

##### 파일 한 줄 읽기
###### 파일을 열어 첫 번째 줄만 출력하시오.
###### 파일이 없으면 예외 처리.

```python
try:
    with open('data.txt','r',encoding='utf-8') as f:
        print(f.readline())
except FileNotFoundError:
    print("파일이 없음")
```

##### 파일 전체 줄 출력
###### 파일 내용을 한 줄씩 출력하시오.
###### 파일이 없으면 예외 처리.


```python
try:
    with open('data.txt','r',encoding='utf-8') as f:
        for i in f:
            print(i.strip())
            
except FileNotFoundError:
    print("파일이 없음")
```

##### 사용자 입력 저장
###### 사용자로부터 문자열을 입력받아
###### memo.txt 파일에 저장하시오.


```python
try:
    with open("memo.txt",'a',encoding='utf-8') as f:
        while True:
            txt = input("입력하고 싶은 내용")
            if txt == "end":
                break
            else:
                f.write(txt + '\n')
except FileNotFoundError:
    print("파일이 없음")
```

##### 숫자 저장
###### 숫자를 입력받아 number.txt 파일에 저장하시오.
###### 숫자가 아니면 예외 처리.


```python
try:
    with open('number.txt','a',encoding='utf-8') as f:
        while True:
            try:
                number = int(input("숫자를 입력"))
                f.write(str(number) + '\n')
            except ValueError:
                break
except FileNotFoundError:
    print("파일이 없음")
```

##### 여러 줄 저장
###### 사용자에게 문자열을 3번 입력받아
###### 파일에 한 줄씩 저장하시오.

```python
try:
    while True:
        with open('list.txt','a',encoding='utf-8') as f:
            text = []
            for t in range(3):
                t = input("문자열 입력")
                if t == "끝":
                    exit()
                text.append(t)
            for i in text:
                f.write(i + '\n')
            text.clear()
except FileNotFoundError:
    print("파일이 없음")
```
